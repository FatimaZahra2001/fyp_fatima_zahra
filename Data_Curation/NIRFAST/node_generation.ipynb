NIRFAST helper functions.

```python
import math
import math
import matplotlib.pyplot as plt
```

```python
def generate_circle_nodes(radius, num_nodes):
    """
    Generate nodes around the circumference of a circle.

    Parameters:
    - radius (float): Radius of the circle in mm.
    - num_nodes (int): Number of nodes to generate.

    Returns:
    - List of tuples: Coordinates of the generated nodes.
    """
    nodes = []
    for i in range(num_nodes):
        angle = (2 * math.pi * i) / num_nodes
        x = radius * math.cos(angle)
        y = radius * math.sin(angle)
        nodes.append((x, y))
    return nodes

def print_nodes(nodes, label="Node"):
    """
    Print the coordinates of nodes without brackets or writing.

    Parameters:
    - nodes (List of tuples): Coordinates of nodes.
    - label (str): Label for the nodes.
    """
    for i, (x, y) in enumerate(nodes):
        # print(f"{label} {i + 1}: {x:.2f} mm, {y:.2f} mm")
        print(f"{i+1} {x:.2f} {y:.2f}")
        

def visualize_circle_and_nodes(radius, nodes, label="Circle"):
    """
    Visualize the circle and nodes using matplotlib.

    Parameters:
    - radius (float): Radius of the circle in mm.
    - nodes (List of tuples): Coordinates of nodes.
    - label (str): Label for the circle.
    """
    print(f"\n{label} Nodes:")
    print_nodes(nodes, label=label)
    
    circle = plt.Circle((0, 0), radius, edgecolor='b', facecolor='none', linestyle='dashed', label=label)
    plt.gca().add_patch(circle)

    node_x, node_y = zip(*nodes)
    plt.scatter(node_x, node_y, color='red', label='Nodes')

    plt.xlabel('X-axis (mm)')
    plt.ylabel('Y-axis (mm)')
    plt.title(f'{label} with Nodes')
    plt.legend()
    plt.axis('equal')  # Ensure the aspect ratio is equal
    plt.show()
```

```python




# Example usage:
radius_circle1 = 45  # Radius of the first circle in mm
radius_circle2 = 35  # Radius of the second circle in mm
num_nodes = 4  # Number of nodes

nodes_circle1 = generate_circle_nodes(radius_circle1, num_nodes)
nodes_circle2 = generate_circle_nodes(radius_circle2, num_nodes)

visualize_circle_and_nodes(radius_circle1, nodes_circle1, label='Circle 1')
visualize_circle_and_nodes(radius_circle2, nodes_circle2, label='Circle 2')
```

```python
def generate_detector_source_pattern(num_sources, num_detectors):
    pattern = []
    for source in range(1, num_sources + 1):
        for detector in range(1, num_detectors + 1):
            if source != detector:
                pattern.append(f"{source} {detector} 1")
    return pattern

def save_pattern_to_file(pattern, filename="output.txt"):
    with open(filename, "w") as file:
        for line in pattern:
            file.write(line + "\n")

# Example usage:
num_sources = 4
num_detectors = 4

pattern = generate_detector_source_pattern(num_sources, num_detectors)
save_pattern_to_file(pattern, filename="output.txt")
```
